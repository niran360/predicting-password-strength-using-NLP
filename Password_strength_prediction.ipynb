# shuffle data to allow for robustness

```python
import random
random.shuffle(password_tuple)
```

```python
x=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]
```

```python
x
```

# prepocessing of text for tfidf

```python
def word_divide_char(inputs):
    character = []
    for i in inputs:
        character.append(i)
    return character
```

```python
vectorizing text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer = word_divide_char)
```

```python
X=vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
init_doc_vector = X[0]
init_doc_vector
```

```python
init_doc_vector.T.todense()
```

```python
df_1 = pd.DataFrame(init_doc_vector.T.todense(), index =vectorizer.get_feature_names(), columns = ['TF-IDF'])
df_1.sort_values(by=['TF-IDF'],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0, multi_class='multinomial')
```

```python
clf.fit(X_train, y_train)
```

# checking a prediction for a sample data

```python
dt = np.array(['$@#34rfgte.'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

# Checking Accuracy of Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")
print("model accuracy is:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
